```python
# collaboration
import numpy
import pandas
import scipy.sparse
n_authors = 5
n_papers = 10
authorship_matrix = pandas.DataFrame(index=[f'author_{i}' for i in range(n_authors)],
                                     columns=[f'paper_{i}' for i in range(n_papers)],
                                     data=numpy.random.randint(low=0, high=2, size=(n_authors, n_papers), dtype=int))
authorship_matrix
```

```python
collaboration_matrix = authorship_matrix.dot(authorship_matrix.T)
collaboration_matrix
```

```python
# neighbor average
n_users = 5
n_items = 10
following_matrix = numpy.random.randint(low=0, high=2, size=(n_users, n_users), dtype=int)
numpy.fill_diagonal(following_matrix, 0)
following_matrix = pandas.DataFrame(index=[f'user_{i}' for i in range(n_users)],
                                     columns=[f'user_{i}' for i in range(n_users)],
                                     data=following_matrix)
rating_matrix = pandas.DataFrame(index=[f'user_{i}' for i in range(n_users)],
                                     columns=[f'item_{i}' for i in range(n_items)],
                                     data=numpy.random.randint(low=0, high=6, size=(n_users, n_items), dtype=int))
rating_matrix
```

```python
following_matrix
```

```python
following_matrix.dot(rating_matrix) / following_matrix.dot(rating_matrix > 0)
```
